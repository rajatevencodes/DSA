```python
class Solution:
    def findMaximumCookieStudents(self, Student, Cookie):
        counter = 0
        Student.sort()
        Cookie.sort()
        i = 0
        j = 0
        while i < len(Cookie) and j < len(Student):
            if Cookie[i] >= Student[j]:
                counter += 1
                i += 1
                j += 1
            elif Cookie[i] < Student[j]:
                i += 1

        return counter


if __name__ == "__main__":
    Student = [
        25369,
        8925,
        30358,
        23116,
        2745,
        15924,
        3882,
        27905,
        26757,
        24913,
        30101,
        15045,
        11121,
        1576,
        26500,
        18395,
        20163,
        25266,
        2973,
        1699,
        2157,
        28144,
        29297,
        9406,
        29082,
        13870,
        9132,
        13171,
        7559,
        27781,
        2959,
        12113,
        4756,
        18851,
        23713,
        11030,
        22939,
        12239,
        14753,
        8656,
        6016,
        11815,
        15294,
        13378,
        25865,
        18732,
        10507,
        21897,
        22768,
        11393,
        28461,
        11008,
        30775,
        10736,
        13910,
        9056,
        188,
        10670,
        8312,
        2348,
        16909,
        28640,
        2226,
        5196,
        30845,
        23389,
        22287,
        16119,
        24184,
        438,
        15471,
        25047,
        26643,
        19572,
        27333,
    ]
    Cookie = [
        23400,
        23783,
        31648,
        7555,
        18082,
        17440,
        12681,
        1619,
        1239,
        22225,
        27585,
        29192,
        7902,
        27787,
        3433,
        14804,
        16753,
        5677,
        24238,
        21691,
        22740,
        23293,
        9620,
        7495,
        4023,
        19448,
        15347,
        8062,
        13662,
        24935,
        18110,
        785,
        20223,
        22195,
        29323,
        26070,
        22636,
        30842,
        22864,
        14968,
        30705,
        6370,
        30059,
        32622,
        10624,
        7726,
        17894,
        25129,
        9483,
        24463,
        2040,
        8850,
        1074,
        17468,
        10703,
        19123,
        27422,
        22724,
        23180,
        3764,
        14509,
        20131,
        23440,
        26890,
        463,
        6676,
        30454,
        18155,
        23600,
        8743,
        16224,
        22789,
        15383,
        14099,
        31991,
        23663,
        27022,
        17101,
        27565,
        28640,
        20675,
    ]
    solution = Solution()
    result = solution.findMaximumCookieStudents(Student, Cookie)
    print("Number of students satisfied:", result) # Expected output: 74
```

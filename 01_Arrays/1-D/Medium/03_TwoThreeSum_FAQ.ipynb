```python
# Brute Force
# Note : Two-pointer approach doesn't work here since sorting would change the original indices

class Solution:
    def twoSum(self, nums, target):
        return self.result(nums, target)

    def result(self, nums, target):
        list = []
        for i in range(len(nums)):
            for j in range(len(nums)):
                if nums[i] + nums[j] == target:
                    list.append(i)
                    list.append(j)
                    return list


# Example usage
if __name__ == "__main__":
    arr = [1, 3, 5, -7, 6, -3]
    target = 0
    sol = Solution()
    result = sol.twoSum(arr, target)
    print(result)
```

```python
# The two-pointer approach is applicable here because we only need to find the numbers, not their indices.
class Solution:
    def threeSum(self, nums):
        return self.result(nums)

    def result(self, nums):

        nums.sort()
        target = 0
        ans = []
        for k in range(len(nums) - 2):

            # Skip duplicate values for k
            if k > 0 and nums[k] == nums[k - 1]:
                continue

            i = k + 1
            j = len(nums) - 1
            while i < j:
                sum = nums[i] + nums[j] + nums[k]
                if sum == target:

                    ans.append([nums[k], nums[i], nums[j]])

                    # Move i and j, but skip duplicates
                    while i < j and nums[i] == nums[i + 1]:
                        i += 1
                    while i < j and nums[j] == nums[j - 1]:
                        j -= 1

                    i += 1
                    j -= 1
                elif sum > target:
                    j -= 1
                elif sum < target:
                    i += 1
        return ans


# Example usage
if __name__ == "__main__":
    arr = [2, -1, -1, 3, -1]
    # target = 0
    sol = Solution()
    result = sol.threeSum(arr)
    print(result)
```

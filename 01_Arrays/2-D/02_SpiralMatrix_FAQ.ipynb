```python
# Optimal Solution

"""Question is simple If the greater to any one to its right then it's a leader"""

""" Time  - O(n) : Space - O(n)  """


class Solution:
    def spiralOrder(self, matrix):
        return self.result(matrix)

    def result(self, matrix):
        minRow = 0
        maxRow = len(matrix) - 1
        minCol = 0
        maxCol = len(matrix[0]) - 1

        total = len(matrix) * len(matrix[0])
        count = 0
        result = []

        while count < total:
            # Traverse right (->)
            for i in range(minCol, maxCol + 1):
                if count < total:
                    result.append(matrix[minRow][i])
                    count += 1
            minRow += 1

            # Traverse down (⬇)
            for i in range(minRow, maxRow + 1):
                if count < total:
                    result.append(matrix[i][maxCol])
                    count += 1
            maxCol -= 1

            # Traverse left (⬅)
            for i in range(maxCol, minCol - 1, -1):
                if count < total:
                    result.append(matrix[maxRow][i])
                    count += 1
            maxRow -= 1

            # Traverse up (⬆)
            for i in range(maxRow, minRow - 1, -1):
                if count < total:
                    result.append(matrix[i][minCol])
                    count += 1
            minCol += 1

        return result


# Example usage
if __name__ == "__main__":
    arr = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
    sol = Solution()
    result = sol.spiralOrder(arr)
    print(result)
```
